```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv('/content/Position_Salaries.csv')
df.head()
```

```python
x= df.iloc[:,1:2].values
y= df.iloc[:, 2].values
```

```python
#fitting Decision Tree on the dataset
from sklearn.tree import DecisionTreeRegressor
regressor= DecisionTreeRegressor(random_state= 0)
regressor.fit(x, y)
```

```python
y_pred= regressor.predict([[6.5]])
y_pred
```

```python
#plotting the Decision Tree Results
plt.scatter(x, y, color='red')
plt.plot(x, regressor.predict(x), color='blue')
plt.title('truth or bluff (decision tree)')
plt.xlabel('position level')
plt.ylabel('salary')
plt.show()
```

```python
#visulaizing the decision tree regression results (for higher resolution and smoother curve)
x_grid= np.arange(min(x), max(x), 0.01)
x_grid= x_grid.reshape((len(x_grid), 1))
plt.scatter(x, y, color='red')
plt.plot(x_grid, regressor.predict(x_grid), color='green')
plt.title('truth or bluff (decision tree)')
plt.xlabel('position level')
plt.ylabel('salary')
plt.show()
```

**Random** **Forest**

```python
#fitting Random forest on the dataset
from sklearn.ensemble import RandomForestRegressor
regressor2= RandomForestRegressor(n_estimators=10, random_state=0)
regressor2.fit(x, y)
```

```python
#visulaizing the random forect regression results (for higher resolution and smoother curve)
x_grid= np.arange(min(x), max(x), 0.01)
x_grid= x_grid.reshape((len(x_grid), 1))
plt.scatter(x, y, color='red')
plt.plot(x_grid, regressor2.predict(x_grid), color='purple')
plt.title('truth or bluff (decision tree)')
plt.xlabel('position level')
plt.ylabel('salary')
plt.show()
```
